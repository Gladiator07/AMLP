## MNIST Visualization

#### Converting supervised data to unsupervised for basic visualizations

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn import datasets
from sklearn import manifold

%matplotlib inline
```

```python
data = datasets.fetch_openml(
                    'mnist_784',
                    version=1,
                    return_X_y=True)

pixel_values, targets = data
targets = targets.astype(int)
```

Reshaping the array size to visualize a single image

```python
single_image = pixel_values[1, :].reshape(28, 28)
plt.imshow(single_image, cmap='gray')
```

```python
tsne = manifold.TSNE(n_components=2, random_state=42)
transformed_data = tsne.fit_transform(pixel_values[:3000, :])
```

TSNE docs [here](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html)

Transformed data is in the shape (3000, 2)

```python
tsne_df = pd.DataFrame(np.column_stack((transformed_data, targets[:3000])),
                      columns=["x", "y", "targets"])
```

x and y are two components from t-SNE decomposition and targets is the actual number

```python
# converting targets (in str) to int
tsne_df.loc[:, "targets"] = tsne_df.targets.astype(int)
```

```python
# first 10 rows of dataframe
tsne_df.head(10)
```

```python
grid = sns.FacetGrid(tsne_df, hue="targets", height=8)
grid.map(plt.scatter, "x", "y").add_legend()
```
